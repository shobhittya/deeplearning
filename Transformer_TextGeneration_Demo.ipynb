```python
# 2. Import libraries
from transformers import pipeline
```

```python
# 3. Load the text-generation pipeline using a lightweight model
text_generator = pipeline("text-generation", model="distilgpt2")
```

```python
# 4. Define a prompt
prompt = "Once upon a time in a future powered by AI,"
```

```python
# 5. Generate text
output = text_generator(prompt, max_length=1000, do_sample=True, temperature=0.8)
```

```python
# 6. Display the result
print("\nGenerated Text:\n")
print(output[0]['generated_text'])
```
